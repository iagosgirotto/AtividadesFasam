```python
import pandas as pd
from sklearn import metrics
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
#Exemplo 1
a, b = make_blobs(n_samples=1800, centers = 4, n_features = 2, random_state = 20)
plt.scatter(a[:, 0], a[:, 1]);
```

```python
#Exemplo 2
n_clusters = range(1, 10)

kmeans = [KMeans(n_clusters=i) for i in n_clusters]
score = [kmeans[i].fit(a).score(a) for i in range(len(kmeans))]

plt.plot(n_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
#Exemplo 3
sil = []
kmax = 10

for k in range(2, kmax+1):
    kmeans = KMeans(n_clusters=k).fit(a)
    labels = kmeans.labels_
    sil.append(metrics.silhouette_score(a, labels, metric = 'euclidean'))

plt.plot(range(2, 11), sil)
plt.xlabel('Numer of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
#Exemplo 4
n_clusters = range(1, 10)

kmeans_ = [KMeans(n_clusters=i) for i in n_clusters]
score = [kmeans_[i].fit(a).score(a) for i in range(len(kmeans_))]
kmeans_[3].predict([[2,8], [-8,4], [8,6], [-2,0]])
plt.scatter(a[:, 0], a[:, 1]);
```
